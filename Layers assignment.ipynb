```python
def is_leap(year):
    leap = False
    if(year%4==0):
        if(year%100==0):
            if(year%400==0):
                leap = True
    
    return leap

year = int(input())
print(is_leap(year))
```

```python
if __name__ == '__main__':
    n = []
    s = []
    for _ in range(int(input())):
        name = input()
        n.append(name)
        score = float(input())
        s.append(score)

    sor = list(set(s))
    sor.sort()
    sl = sor[1]
    o = []
    for x, i in enumerate(s):
        if sl == i:
            o.append(n[x])
    o.sort()
    print(*o, sep = "\n")
```

```python
def average(a):
    return sum(a)/len(a)
if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    print('%.2f'%average(student_marks[query_name]))
```
